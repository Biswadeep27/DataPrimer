```python
#1422: Maximum Score After Splitting a String
class Solution:
    def maxScore(self, s: str) -> int:

        zeros = 0
        ones = s.count('1')
        score = 0

        for i in range(len(s) - 1):
            if s[i] == '0':
                zeros += 1
            else:
                ones -= 1

            score = max(score, zeros + ones)

        return score
```

```python
#189. Rotate Array
# with Space Complexity O(n)
class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        tem_arr = [0]* len(nums)

        for i in range(len(nums)):
            idx = (i + k) % len(nums) 
            tem_arr[idx] = nums[i]

        nums[:] = tem_arr
```

```python
# with space complexity O(1)
class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        k = k % len(nums) # for cases where k is karger than len(nums)
        def list_reverse(arr : List[int], l: int, r: int):
            while l < r :
                arr[l], arr[r] = arr[r], arr[l]
                l += 1
                r -= 1
        
        list_reverse(nums, 0, len(nums) - 1 ) #O(n/2)
        list_reverse(nums, 0 , k - 1) #O(k/2)
        list_reverse(nums, k, len(nums) - 1) #O(n-k/2)
```

```python
#125: Valid Palindrome
class Solution:
    def isPalindrome(self, s: str) -> bool:
        
        new_str = ''
        for c in s:
            new_str += c.lower() if c.isalnum() else ''

        l = 0
        r = len(new_str) - 1

        while l <= r:
            if new_str[l] != new_str[r]:
                return False

            l, r = l + 1, r - 1

        return True
```

```python
#1758. Minimum Changes To Make Alternating Binary String

class Solution:
    def minOperations(self, s: str) -> int:
        # either we need to get to a1 = 010101.. or a2 = 101010.. : need min altercation
        # logic : if you need c moves to get to a1 then you would need len(s) - c 
        # moves to get to a2.

        count = 0

        for i in range(len(s)):
            count = count + 1 if i % 2 == 0 and s[i] != '0' else count + 1 if i % 2 != 0 and s[i] != '1' else count

        return min(count, len(s) - count)
```

```python
#15: three sum
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        res = []

        nums.sort()

        for i,num in enumerate(nums):
            if i > 0 and nums[i] == nums[i-1]:
                i += 0
                continue

            l,r = i + 1, len(nums) - 1
            while l < r:
                three_sum = num + nums[l] + nums[r]
                if three_sum > 0:
                    r -= 1
                elif three_sum <0:
                    l += 1
                else:
                    res.append([num , nums[l] , nums[r]])
                    l += 1 # next 2-sum comb withot repetition
                    while nums[l] == nums[l-1] and l < r:
                        l += 1

        return res
```

```python
#91. Decode Ways
# recursion with memoization
class Solution:
    def numDecodings(self, s: str) -> int:
        # recursion - memoization : O(n) time and space
        dp = {len(s) : 1}

        def dfs(i):
            if i in dp:
                return dp[i]
            if s[i] == "0":
                return 0

            res = dfs(i + 1) #considering s[i] as valid single digit for decoding
            if (i+1 < len(s) and (s[i] == "1" or s[i] == "2" and s[i+1] in "0123456")):
                res += dfs(i+2) # #considering s[i+1] as valid double digit for decoding

            dp[i] = res
            return res

        return dfs(0)
```

```python
#1155. Number of Dice Rolls With Target Sum
class Solution:
    def numRollsToTarget(self, n: int, k: int, target: int) -> int:
        dp = {}
        MOD = 10**9 + 7
        def count_ways(n,target):
            #global ways
            if (n,target) in dp.keys():
                return dp[(n,target)]
            if n == 0:
                return 1 if target == 0 else 0
            if target < 0:
                return 0
            ways = 0
            for i in range(1,k+1):
                ways += count_ways(n-1, target - i)

            dp[(n,target)] = ways
            return ways

        return count_ways(n, target) % MOD
```

```python
# dp
class Solution:
    def numRollsToTarget(self, n: int, k: int, target: int) -> int:
        # fun problem
        
        # unique dice, e.g. 6+1 and 1+6 are different ways

        # brute-force: top-down DP:
        #      ways(n,k) =
        #        ways(n-1,k-1) # roll a one, then continue
        #       +ways(n-2,k-1) # roll a two, then continue
        #       +...
        #       +ways(n-k,k-1)

        # So ways(m,k) only depends on k-1: in-place DP trick to reduce memory use

        # Maybe faster: for 1..k we can roll
        #    r_k rolls of k: have to get target-r_k*k with rolls of 1..k-1 in k-r_k
        #    then we get r_1 .. r_k: can use combinatorics to get the total ways
        #    but this is really just the top-down DP way

        if target < n: return 0 # min roll is 1
        if target == n: return 1 # min roll every time
        if target == n*k: return 1 # max roll every time
        if target > n*k: return 0

        ### DP solution: will be faster for cumulative sum reasons, and less memory
        # O(target) work per iteration, O(n) iterations: O(n*target)
        ways = [0]*(target+1)
        ways[0] = 1
        BIG = 10**9+7
        for _ in range(n): # exactly k rolls
            w = sum(ways[t] for t in range(max(target-k, 0),target)) % BIG # sum of ways to roll prior k values
            ways[target] = w
            for t in range(target-1, -1, -1): # TODO: can optimize b/c we know leading m < n entries are 0
                if t-k >= 0: w += ways[t-k]
                w -= ways[t]
                ways[t] = w

        return ways[target] % BIG
```

```python
# 121. Best Time to Buy and Sell Stock
# sliding window
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        l, r = 0, 1
        max_profit = 0

        while r < len(prices):
            if prices[l] < prices[r]:
                profit = prices[r] - prices[l]
                max_profit = max(max_profit, profit)
            else:
                l = r

            r += 1

        return max_profit
```

```python
# 3. Longest Substring Without Repeating Characters
# hashSet and sliding window
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        charSet = set()
        l = 0
        res = 0

        for r in range(len(s)):
            while s[r] in charSet:
                charSet.remove(s[l])
                l += 1
            charSet.add(s[r])
            res = max(res, r - l + 1)

        return res
```

```python
# 209. Minimum Size Subarray Sum
class Solution:
    def minSubArrayLen(self, target: int, nums: List[int]) -> int:
        l = prefix_sum = 0
        res = float("inf")

        for r in range(len(nums)):
            prefix_sum += nums[r]
            while prefix_sum >= target:
                res = min(res, r - l + 1)
                prefix_sum -= nums[l]
                l += 1

        return 0 if res == float("inf") else res
```

```python
# 424. Longest Repeating Character Replacement
# hash_map and sliding window : 0(26*n)
class Solution:
    def characterReplacement(self, s: str, k: int) -> int:
        hash_map = {}
        l = res = 0

        for r in range(len(s)):
            hash_map[s[r]] = hash_map.get(s[r],0) + 1
            len_substring = r - l + 1
            while len_substring - max(hash_map.values()) > k:
                hash_map[s[l]] -= 1
                l += 1
                len_substring = r - l + 1

            res = max(res, len_substring)

        return res
```

```python
# 424. Longest Repeating Character Replacement
# hash_map and sliding window : 0(n) -> removing the max fn on dict values 
class Solution:
    def characterReplacement(self, s: str, k: int) -> int:
        hash_map = {}
        l = res = maxf = 0
        
        for r in range(len(s)):
            hash_map[s[r]] = hash_map.get(s[r],0) + 1
            maxf = max(maxf, hash_map[s[r]])
            len_substring = r - l + 1
            while len_substring - maxf > k:
                hash_map[s[l]] -= 1
                l += 1
                len_substring = r - l + 1

            res = max(res, len_substring)

        return res
```

```python
# 1531. String Compression II
# recursion with memoization
class Solution:
    def getLengthOfOptimalCompression(self, s: str, k: int) -> int:
        cache = {}

        def count(i,k,prev,prev_count):
            if (i,k,prev,prev_count) in cache:
                return cache[(i,k,prev,prev_count)]
            if k < 0:
                return float("inf")
            if i == len(s):
                return 0

            if s[i] == prev:
                incr = 1 if prev_count in [1,9,99] else 0
                res = incr + count(i+1,k,prev, prev_count + 1) # will continue the streak without deleting
            else:
                res = min(
                    count(i+1,k-1,prev,prev_count), # deleting s[i]
                    1 + count(i+1,k,s[i],1) # not deleting s[i]
                )

            cache[(i,k,prev,prev_count)] = res
            return res
        return count(0,k,"",0)
```

```python
#567. Permutation in String
# O(26*n) -> O(26) + O(n)
class Solution:
    def checkInclusion(self, s1: str, s2: str) -> bool:
        if len(s1) > len(s2):
            return False
        
        s1Count, s2Count = 26 * [0], 26 * [0]

        for i in range(len(s1)):
            s1Count[ord(s1[i]) - ord('a')] += 1
            s2Count[ord(s2[i]) - ord('a')] += 1

        matches = 0
        for i in range(26):
            matches += (1 if s1Count[i] == s2Count[i] else 0)

        l = 0
        for r in range(len(s1),len(s2)):
            if matches == 26:
                return True

            index = ord(s2[r]) - ord('a')
            s2Count[index] += 1
            if s1Count[index] == s2Count[index]:
                matches += 1
            elif s1Count[index] + 1 == s2Count[index]:
                matches -= 1

            index = ord(s2[l]) - ord('a')
            s2Count[index] -= 1
            if s1Count[index] == s2Count[index]:
                matches += 1
            elif s1Count[index] - 1 == s2Count[index]:
                matches -= 1

            l += 1

        return matches == 26
```

```python
# 1335. Minimum Difficulty of a Job Schedule
class Solution:
    def minDifficulty(self, jobDifficulty: List[int], d: int) -> int:
        if len(jobDifficulty) < d:
            return -1

        cache = {}
        def dfs(i,d,cur_max):
            if i == len(jobDifficulty):
                return 0 if d == 0 else float('inf')
            if d == 0:
                return float('inf')

            if (i,d,cur_max) in cache:
                return cache[(i,d,cur_max)]
            
            cur_max = max(cur_max, jobDifficulty[i])

            res = min(
                dfs(i+1,d,cur_max), #continue day
                cur_max + dfs(i+1,d-1,-1) #end day
            )
            cache[(i,d,cur_max)] = res
            return res

        return dfs(0,d,-1)
```

```python
#20. Valid Parentheses
#stack : O(n)
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        valid = {')':'(','}':'{',']':'['}

        for c in s:
            if c not in valid:
                stack.append(c)
            else:
                if len(stack) > 0 and valid[c] == stack[-1]:
                    stack.pop()
                else:
                    return False

        return len(stack) == 0
```

```python
#155. Min Stack
#O(1) : at each stack push, we calculae the corresponding min, so when we pop to go back to the previous state, we just remove the corresponding min of
#that particular state to check the min from before push of the popped element.
class MinStack:

    def __init__(self):
        self.stack = []
        self.minStack = []

    def push(self, val: int) -> None:
        self.stack.append(val)
        min_val = min(val, self.minStack[-1] if self.minStack else val)
        self.minStack.append(min_val)

    def pop(self) -> None:
        self.minStack.pop()
        self.stack.pop()

    def top(self) -> int:
        return self.stack[-1]

    def getMin(self) -> int:
        return self.minStack[-1]
```

```python
# 739. Daily Temperatures
class Solution:
    def dailyTemperatures(self, temperatures: List[int]) -> List[int]:
        res = [0] * len(temperatures)
        stack = []

        for i,t in enumerate(temperatures):
            while stack and t > stack[-1][0]:
                stackT, stackInd = stack.pop()
                res[stackInd] = i - stackInd

            stack.append((t,i))

        return res
```

```python
# 2610. Convert an Array Into a 2D Array With Conditions
from collections import defaultdict
class Solution:
    def findMatrix(self, nums: List[int]) -> List[List[int]]:
        res = []
        count = defaultdict(int)

        for n in nums:
            row = count[n]
            if len(res) == row:
                res.append([])

            res[row].append(n)
            count[n] += 1

        return res
```

```python
# 74. Search a 2D Matrix
class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        top = l = 0
        bottom = len(matrix) - 1
        r = len(matrix[0]) - 1

        while top <= bottom:
            row = (top + bottom) // 2
            if target > matrix[row][-1]:
                top = row + 1
            elif target < matrix[row][0]:
                bottom = row - 1
            else:
                break

        if not top <= bottom:
            return False

        row = (top + bottom)//2
        while l<=r:
            mid = (l + r)//2
            if target > matrix[row][mid]:
                l = mid + 1
            elif target < matrix[row][mid]:
                r = mid - 1

            else:
                return True

        return False
```

```python
# 875. Koko Eating Bananas
# binary search
class Solution:
    def minEatingSpeed(self, piles: List[int], h: int) -> int:
        l = 1
        r = res = max(piles)

        while l <= r:
            mid = (l + r)//2
            time_taken = 0
            for p in piles:
                time_taken += (p // mid) + (p % mid > 0)
            if time_taken <= h:
                res = min(res, mid)
                r = mid - 1
            else:
                l = mid + 1

        return res
```

```python
# 2870. Minimum Number of Operations to Make Array Empty
# with recursion and memoization
class Solution:
    def minOperations(self, nums: List[int]) -> int:
        cache = {}
        def dfs(n):
            if n < 0:
                return float("inf")
            if n in [2, 3]:
                return 1
            if n in cache:
                return cache[n]

            res = min(dfs(n-2), dfs(n-3))
            # if res == -1:
            #     return -1
            res += 1
            cache[n] = res
            return res

        count = Counter(nums)
        res = 0
        for n,c in count.items():
            op = dfs(c)
            if op == float("inf"):
                return -1
            res += op

        return res
```

```python
# 300. Longest Increasing Subsequence:
# Brute Force : O(2^n)
# DP : O(n^2)
class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        dp = [1] * len(nums)

        for i in range(len(nums)-1,-1,-1):
            dp[i] = 1
            for j in range(i+1,len(nums)):
                if nums[i] < nums[j]:
                    dp[i] = max(dp[i], 1 + dp[j])

        return max(dp)
```

```python
# 1143. Longest Common Subsequence
# 2D-DP : O(n*m)
class Solution:
    def longestCommonSubsequence(self, text1: str, text2: str) ->int:
        dp = [[0 for j in range(len(text2)+1)] for i in range(len(text1)+1)]

        for i in range(len(text1)-1,-1,-1):
            for j in range(len(text2)-1,-1,-1):
                if text1[i] == text2[j]:
                    dp[i][j] = 1 + dp[i+1][j+1]
                else:
                    dp[i][j] = max(dp[i+1][j], dp[i][j+1])

        return dp[0][0]
```

```python
# 5. Longest Palindromic Substring
# 2 pointer : O(n^2)
class Solution:
    def longestPalindrome(self, s: str) -> str:
        res = ""
        resLen = 0

        for i in range(len(s)):
            l = r = i
            while l >= 0 and r < len(s) and s[l] == s[r]:
                if r - l + 1 > resLen:
                    res = s[l:r+1]
                    resLen = r - l + 1

                l -= 1
                r += 1

            l = i
            r = i + 1
            while l >= 0 and r < len(s) and s[l] == s[r]:
                if r - l + 1 > resLen:
                    res = s[l:r+1]
                    resLen = r - l + 1

                l -= 1
                r += 1

        return res
```
